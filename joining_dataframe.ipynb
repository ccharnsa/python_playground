```python
# https://www.datacamp.com/community/tutorials/joining-dataframes-pandas


import pandas as pd

##### DF1

dummy_data1 = {
        'id': ['1', '2', '3', '4', '5'],
        'Feature1': ['A', 'C', 'E', 'G', 'I'],
        'Feature2': ['B', 'D', 'F', 'H', 'J']}

df1 = pd.DataFrame(dummy_data1, columns = ['id', 'Feature1', 'Feature2'])

df1



## OR

# pieces = {'x': df1, 'y': df2}
# df_piece = pd.concat(pieces)
```

```python
##### DF2
dummy_data2 = {
        'id': ['1', '2', '6', '7', '8'],
        'Feature1': ['K', 'M', 'O', 'Q', 'S'],
        'Feature2': ['L', 'N', 'P', 'R', 'T']}

df2 = pd.DataFrame(dummy_data2, columns = ['id', 'Feature1', 'Feature2'])

df2
```

```python
##### DF3
dummy_data3 = {
        'id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'Feature3': [12, 13, 14, 15, 16, 17, 15, 12, 13, 23]}

df3 = pd.DataFrame(dummy_data3, columns = ['id', 'Feature3'])

df3
```

```python
#### concatenate - not ignore index

df_row = pd.concat([df1, df2])
df_row
```

```python
#### concatenate - ignore index

df_row_reindex = pd.concat([df1, df2], ignore_index=True)

df_row_reindex
```

```python
#### concatenate - add label

frames = [df1,df2]
df_keys = pd.concat(frames, keys=['x', 'y'])

df_keys

# get only specific key
df_keys.loc['y']
```

```python
#### Merge DataFrames, number of row has to be the same

df_merge_col = pd.merge(df_row, df3, on='id')

df_merge_col
```

```python
#### append row
add_row = pd.Series(['10', 'X1', 'X2', 'X3'],
                    index=['id','Feature1', 'Feature2', 'Feature3'])

df_add_row = df_merge_col.append(add_row, ignore_index=True)

df_add_row
```

```python
##### outer join

# The joined DataFrame will contain all records from both the DataFrames and fill in NaNs for missing matches on either side. 


df_outer = pd.merge(df1, df2, on='id', how='outer')

df_outer
```

```python
###### outer join with suffix
df_suffix = pd.merge(df1, df2, left_on='id',right_on='id',how='outer',suffixes=('_left','_right'))

df_suffix
```

```python
##### inner join

# produces only the set of records that match in both DataFrame A and DataFrame B. 

df_inner = pd.merge(df1, df2, on='id', how='inner')

df_inner
```

```python
#### right join

# The RIGHT JOIN produces a complete set of records from DataFrame B (right DataFrame), with the matching records (where available) in DataFrame A (left DataFrame). If there is no match, the right side will contain null. 

df_right = pd.merge(df1, df2, on='id', how='right')

df_right
```

```python
#### left join

# The LEFT JOIN produces a complete set of records from DataFrame A (left DataFrame), with the matching records (where available) in DataFrame B (right DataFrame). If there is no match, the left side will contain null.

df_left = pd.merge(df1, df2, on='id', how='left')

df_left
```

```python
##### Joining on index

# Sometimes you may have to perform the join on the indexes or the row labels. To do so, you have to specify right_index (for the indexes of the right DataFrame) and left_index (for the indexes of the left DataFrame) as True :

df_index = pd.merge(df1, df2, right_index=True, left_index=True)

df_index
```
